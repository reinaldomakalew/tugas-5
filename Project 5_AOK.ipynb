## Informasi Mengenai Data 
Data yang digunakan merupakan data survey mengenai memilih hardware komputer 

Pertanyaan-pertanyaan yang diajukan dalam melakukan survey

1. Apakah hardisk tipe ssd memiliki performa kerja lebih baik dari hdd dalam hal kecepatan mengakses data
2. Apakah pengguna setuju bahwa dual core (multi core )pada prosessor lebih cenderung mempercepat kinerja system computer dibandingkan quad core
3. Apakah dengan menggunakan VGA NVIDIA GeForce saat akan melakukan aktivitas gaming tidak ada lag
4. Agar dapat support dengan aplikasi yang bersifat multi core harus menggunakan memori yang besar 
5. Saat melakukan aktivitas multitasking pada komputer  dengan menggunakan processor multi core tidak akan menjadi masalah
6. Apakah pengguna bisa membedakan antara hardware yang masih baru dengan yang sudah second (bekas)
7. Menurut pengalaman pengguna, Apakah tempat service komputer dapat menjamin tidak adanya penukaran hardware computer dengan yang bekas
8. Apakah dengan media penyimpanan yang lebih besar pengguna dapat lega untuk menyimpanan data
9. Pengguna dapat lebih meminimalkan penggunaan listrik jika menggunakan Monitor LCD dari pada monitor CRT

Survey dilakukan kepada 500 responden dengan usia 15-30 tahun(pria dan wanita)

Semakin tinggi nilai koefisien korelasi yang didapat antar sampel data, maka tingkat saling mempengaruhi antar sampel data kuat atau sampel data saling berkaitan. Jika sebaliknya nilai koefisien korelasi yang didapat kecil, maka tingkat saling mempengaruhi antar sampel data lemah atau sampel data tidak saling berkaitan.

### 3. Uji Data menggunakan Principal Component Analysis

```python
mydata<-read.csv('DATAASLI.csv')
```

#### * Menampilkan Data

```python
mydata
```

```python
attach(mydata)
```

```python
#Define Varibel
x <- cbind(Q1,Q2,Q3,Q4,Q5,Q6,Q7,Q8,Q9)
```

#### * Statistik Data
Menampilkan nilai mean, min, max, standard deviasi, correlation dari data

```python
#Descriptive Statistics
summary(x)
cor(x)
```

```python
#Principal Component Analysis
pca <-princomp(x, score=TRUE, cor=TRUE)
```

#### * Principal Component Analysis

```python
#Loading of Principal Components
loadings(pca)
```

```python
#Scree Plot of Eigenvalues
plot(pca)
screeplot(pca, type="line", main="scree plot")
```

#### Pada grafik plot dan screeplot terlihat bahwa komponen satu memiliki grafik tertinggi sehingga harus dipertahankan dan menjadi acuan untuk menentukan nila PCA pada semua sampel data.

```python
#Biplot of Score Variables
biplot(pca)
```

#### Pada grafik biplot terlihat bahwa informasi disederhanakan dari banyak peubah menjadi grafik dua dimensi agar lebih mudah untuk dipahami.

```python
#Scores of The Components
pca$scores[1:10,]
```

#### Pada score PCA memiliki korelasi satu sama lain. score yang didapat bersifat positif dan negatif. Angka yang bercetak tebal menunjukkan korelasi antar  sampel. Jika nilai korelasi antara sampel mencapai nilai 0,5 maka dapat dikatakan sampel tersebut saling berkorelasi, sedangkan jika nilai korelasi antara sampel mencapai 0,8 maka dapat dikatakan sampel tersebut memiliki korelasi yang tinggi.

### Kesimpulan

#### Kesimpulan yang dapat diambil dari pengujian kelayakan data ini ialah agar dapat mengetahui data yang digunakan layak atau tidak untuk digunakan. Nahh dibawah ini merupakan kesimpulan yang telah dirangkum ke dalam beberapa point sebagai berikut.
